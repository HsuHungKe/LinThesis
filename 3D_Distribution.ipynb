```python
import numpy
import matplotlib.pyplot as plt 
```

```python
# Create normalized data points for three dimentions
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
x = numpy.random.normal(7, 2,400)
y = numpy.random.normal(4,2,400)
z = numpy.random.normal(7,2,400)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for i in range(len(x)):
    if 1<x[i]<13 and -2<y[i]<10 and 1<z[i]<13:
        ax.scatter(x[i], y[i], z[i], marker='o', color='b')
        
    else:
        ax.scatter(x[i], y[i], z[i], marker='^', color='r')
plt.show()
```

```python
# Based on above, creating a function
def draw_sphere (mean, var, num):
    x = numpy.random.normal(mean,var,num)
    y = numpy.random.normal(mean,var,num)
    z = numpy.random.normal(mean,var,num)
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(x, y, z, marker='o', color='r')
    plt.show()
```

```python
draw_sphere(6,4,500)
```
